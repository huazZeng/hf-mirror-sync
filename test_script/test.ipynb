```python
import requests

# Part 1: 获取本地公网IP及地理信息，并仅打印 city, region, country 字段
try:
    resp = requests.get("https://ipinfo.io", timeout=5)
    resp.raise_for_status()
    info = resp.json()
    for k in ("city", "region", "country"):
        if k in info:
            print(f'"{k}": "{info[k]}"')
except requests.RequestException as e:
    print(f"Failed to fetch ipinfo.io: {e}")

# Part 2: 发请求模拟 curl -v --max-redirs 0 ...
url = "https://huggingface.co/datasets/nebius/SWE-rebench/resolve/main/README.md"
try:
    r = requests.get(url, allow_redirects=False)
    print(f"Status code: {r.status_code}")
    print("Headers:")
    for k, v in r.headers.items():
        print(f"{k}: {v}")
except requests.RequestException as e:
    print(f"Request failed: {e}")
url = "https://huggingface.co/datasets/nebius/SWE-rebench/resolve/main/data/test-00001-of-00002.parquet"
try:
    r = requests.get(url, allow_redirects=False)
    print(f"Status code: {r.status_code}")
    print("Headers:")
    for k, v in r.headers.items():
        print(f"{k}: {v}")
except requests.RequestException as e:
    print(f"Request failed: {e}")
```
